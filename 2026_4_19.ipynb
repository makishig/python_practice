```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
```

```python
#オブジェクト指向インターフェース
x = [1,2,3]
y = [2,4,9]
fig,ax = plt.subplots() #描画オブジェクトとサブプロットを作成  描画がキャンバス、サブがグラフのイメージ
ax.plot(x,y)
ax.set_title('object-oriented interface')
plt.show()
```

```python
fig,ax = plt.subplots(2) #サブプロットを二つ指定
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1,2,3],[1,2,3], label='Legend label 1') #凡例を表示
ax.plot([1,2,3],[3,2,1], label='legend label 2')
ax.set_xlabel('x label')#軸ラベルを追加
ax.set_ylabel('y label')
ax.legend(loc='best')

fig.savefig('sample-figure.png') #png形式で保存

plt.show()
```

```python
#sin cosの描画
x = np.arange(0.0, 15.0, 0.1)#０から１５まで０．１間隔の一次元データ
y1 = np.sin(x)
y2 = np.cos(x)

fig,ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```
