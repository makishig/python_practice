```python
df = pd.read_excel('data/202411health.xlsx')
df
```

```python
df.loc[:, '歩数'] >= 10000
```

```python
df_selected = df.loc[df.loc[:, '歩数'] >= 10000, :]
df_selected
```

```python
df.query('歩数 >= 10000 and 摂取カロリー <= 1800')
```

```python
df.dtypes  #日付は文字列
```

```python
df['date'] = df.loc[:, '日付'].apply(pd.to_datetime) #カラムを追加、代入
df.loc[:, 'date']
```

```python
df.dtypes
```

```python
df['摂取カロリー'] = df.loc[:, '摂取カロリー'].astype(np.float32)
```

```python
df.head()
```

```python
df = df.set_index('date')
df.head(9)
```

```python
df.sort_values(by='歩数').tail()
```

```python
df = df.drop('日付', axis=1)
```

```python
df.head()
```

```python
df['歩数/カロリー'] = (df['歩数']/df['摂取カロリー'])
df.head()
```

```python
def exercise_judge(ex):
    if ex <= 3.0:
        return 'Low'
    elif ex <= 6.0:
        return 'Mid'
    else:
        return 'High'

df['運動指数'] = df.loc[:, '歩数/カロリー'].apply(exercise_judge)
df.head()
```

```python
df.to_pickle('data/202411health.pickle')
```
