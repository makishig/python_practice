```python
df = df.set_index('date')
df.head(9)
```

```python
df = df.drop('日付', axis=1)
```

```python
df.head()
```

```python
df['歩数/カロリー'] = (df['歩数']/df['摂取カロリー'])
df.head()
```

```python
def exercise_judge(ex):
    if ex <= 3.0:
        return 'Low'
    elif ex <= 6.0:
        return 'Mid'
    else:
        return 'High'

df['運動指数'] = df.loc[:, '歩数/カロリー'].apply(exercise_judge)
df.head()
```

```python
fig, ax = plt.subplots()
x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
plt.show()
```

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']
width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label='y1')
x2 = [num +width for num in x]
ax.bar(x2, y2, width=width, label='y2')
ax.legend()
plt.show()
```

```python
rng = np.random.default_rng(123)
mu = 100  #平均値
x0 = rng.normal(mu, 20, 1000)
x1 = rng.normal(mu, 15, 1000)
x2 = rng.normal(mu, 10, 1000)
fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
ax.hist((x0, x1, x2), label=labels, stacked=True)
ax.legend()
plt.show()
```

```python
x0 = rng.normal(0, 10, 500)
x1 = rng.normal(0, 15, 500)
x2 = rng.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
ax.boxplot((x0, x1, x2), tick_labels=labels)
plt.show()
```

```python
labels = ['spam', 'ham', 'egg']
x = [10, 3, 1]
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.1f%%')
plt.show()
```

```python
explode = [0, 0.2, 0]
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, autopct='%1.2f%%', explode=explode)
plt.show()
```
