```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')

x = [1,2,3]
y = [2,4,9]

fig,ax = plt.subplots()
ax.plot(x,y, label = 'legend label')#折れ線グラフ
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend(loc='best')
plt.show()
```

```python
#棒グラフ
fig, ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]
labels = ['spam','ham','egg']
ax.bar(x,y, tick_label=labels)
plt.show()
```

```python
#複数の棒グラフ
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
label = ['spam', 'ham', 'egg']

width = 0.4     #棒グラフの幅
ax.bar(x, y1, width=width, tick_label=label, label='y1')

x2 = [i + width for i in x]  #幅の長さ分だけずらす
ax.bar(x2, y2, width=width, label='y2')
ax.legend()
plt.show()
```

```python
#散布図
import numpy as np
fig, ax = plt.subplots()
rng = np.random.default_rng(123)
x = rng.random(50)#ランダムに５０個のデータを作成
y = rng.random(50)

ax.scatter(x[:10], y[:10], marker='v', label='sitasannkaku')
ax.scatter(x[10:20], y[10:20], marker='^', label='uesannkaku')
ax.scatter(x[20:30], y[20:30], marker='s', label='sikaku')
ax.scatter(x[30:40], y[30:40], marker='*', label='star')
ax.scatter(x[40:], y[40:], marker='x', label='X')
ax.legend()
plt.show()
```

```python
#ヒストグラムの描画
mu = 100 #平均
sigma = 15 #標準偏差
x = rng.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x) #n は各binの度数　binsはビンの境界の値　
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(x, bins=25)  #ビンの数を指定
plt.show()
```

```python
#複数の値を表示
x0 = rng.normal(mu, 20, 1000)
x1 = rng.normal(mu, 15, 1000)
x2 = rng.normal(mu, 10, 1000)
fig, ax = plt.subplots()

labels = ['x0','x1','x2']
ax.hist((x0, x1, x2),label=labels)
ax.legend()
plt.show()
```

```python
#箱ひげ図
fig, ax = plt.subplots()
ax.boxplot((x0, x1, x2),tick_labels=labels) #データは以前のものを使用
plt.show()
```

```python
#円グラフ
labels = ['spamm','ham','egg']
x = [10, 3, 1]
fig, ax = plt.subplots()

ax.pie(x, labels=labels)
plt.show()                  #デフォルトで三時の位置から反時計回りの順番
```

```python
#十二時の位置から時計回り　影　切り出し　パーセント表示
fig, ax = plt.subplots()
explode = [0, 0.2, 0] #hamを切り出す
ax.pie(x, labels=labels, startangle=90, counterclock=False,
       shadow=True, autopct='%1.2f%%', explode=explode)
plt.show()
```

```python
#複数のグラフを組み合わせる
fig, ax = plt.subplots()
x1 = [1, 2, 3]
y1 = [5, 2, 3]
x2 = [1, 2, 3, 4]
y2 = [8, 5, 4, 6]
ax.bar(x1, y1,label='y1')   #棒グラフ
ax.plot(x2, y2, label='y2')  #折れ線グラフ
ax.legend()
plt.show()
```
