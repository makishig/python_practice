##タイピングゲームのコードを細かく理解する

```python
import pygame   #ゲームを作るためのライブラリ、ウィンドウ表示、キーボード入力、音、画像、描画を行う
import random

pygame.init() #pygameを起動
```

```python
#画面サイズを指定
WIDTH = 800
HEIGHT = 400
screen = pygame.display.set_mode((WIDTH, HEIGHT)) #ウィンドウ作成
pygame.display.set_caption('Typing Game') #ウィンドウ上部のタイトル
WHITE = (255, 255, 255) #RGBで指定
BLACK = (0, 0, 0)

font = pygame.font.Font(None, 74)    #フォントの種類と文字の大きさ
small_font = pygame.font.Font(None, 50)
```

```python
#答え、入力文字
words = [
    "python",
    "apple",
    "banana",
    "keyboard",
    "science",
    "computer"
]
current_word = random.choice(words) #答えをランダムにwordsから選択
typed_text = '' #追加していくから空文字
score = 0
clock = pygame.time.Clock()
```

```python
running = True
while running:  #Trueじゃダメ、後々running = Falseでゲームを終了するから
    #中でやることは、入力確認、画面更新、入力確認....
    screen.fill(WHITE) #毎フレーム画面を白くする。これをしないと前の文字が残り続ける
    
    for event in pygame.event.get(): #イベント取得
        
        if event.type == pygame.QUIT: #もしイベントが終了（×ボタン）なら
            running = False
        
        if event.type == pygame.KEYDOWN: #もしキーが押されたら
            
            if event.key == pygame.K_RETURN: #enter押した？
                
                if typed_text == current_word: #正誤判定
                    score += 1
                    current_word =random.choice(words) #次の単語を設定
                typed_text = '' #入力をリセット
                
            elif event.key == pygame.K_BACKSPACE: #バックスペースの場合
                typed_text = typed_text[:-1] #入力を一つ減らす
                
            else:
                typed_text += event.unicode
    
    #画面の描画
    #フォントの種類、サイズをfontで指定、render()で描画用画像を作る、第一引数で表示する文字、第二引数はアンチエイリアス（文字を滑らかにするもの）、第三で色を指定
    word_surface = font.render(current_word, True, BLACK)  #戻り値には文字画像が入る
                
    screen.blit(word_surface, (250, 100)) #画面に貼る　数字は座標
    
    text_surface = small_font.render(typed_text, True, BLACK) #入力文字の画像を作成
    screen.blit(text_surface, (250, 220))
    
    score_surface = small_font.render(f"Score: {score}", True, BLACK) #スコア表示の画像
    screen.blit(score_surface, (20, 20))
    
    pygame.display.update() #ここで初めて画像更新、画像を反映する
    
    clock.tick(60)

pygame.quit() #pygameの停止
```
