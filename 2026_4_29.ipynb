```python
#主成分分析
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(123)
x = rng.uniform(size=50) #0~1の一様乱数を50個
y = 2*x + 0.5*rng.uniform(size=50)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```
