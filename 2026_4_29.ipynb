scikit-learnのつづき

```python
#主成分分析
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(123)
x = rng.uniform(size=50) #0~1の一様乱数を50個
y = 2*x + 0.5*rng.uniform(size=50)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

```python
from sklearn.decomposition import PCA
#主成分のクラスをインスタンス化
pca = PCA(n_components=2)
x_pca = pca.fit_transform(np.hstack((x.reshape(-1, 1),       #列方向に変換したｘｙを横に連結
                                     y.reshape(-1, 1))))     #x_pca shape=(50,2) ０列目がpc1, １列目がpc2
#可視化
fig, ax = plt.subplots()
ax.scatter(x_pca[:, 0], x_pca[:, 1])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
plt.show()
```
