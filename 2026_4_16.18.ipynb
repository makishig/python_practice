```python
#一次元配列の作成
import numpy as np
a=np.array([1,2,3])
a
```

```python
print(a)
type(a)
```

```python
#二次元配列の作成
b=np.array([[1,2,3],[4,5,6]])
print(b)
type(b)
b.shape
```

```python
#変形 reshape
c1=np.array([0,1,2,3,4,5])
print(c1)
c2=c1.reshape((2,3))
print(c2)
```

```python
c3=c2.ravel()   #ravelは参照を返す
print(c3)
c4=c2.flatten()  #flattenはコピーを返す
c4
```

```python
#データ型について  numpy配列の要素は一つのデータ型に統一される
a.dtype
```

```python
#データ型の宣言、変更
d=np.array([1,2], dtype=np.int16)
d.dtype
```

```python
#astypeメソッドで変換
d.astype(np.float16)
```
