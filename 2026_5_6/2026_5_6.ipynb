##画像データの処理

```python
import os
print(os.getcwd())
from PIL import Image
sample = Image.open('tiger.png')
```

```python
print(type(sample))
sample
```

```python
sample.size
```

```python
print(sample.format)
print(sample.mode)
```

```python
sample.convert('L') #グレースケール
```

```python
sample.show() #widowsのアプリで表示
```

```python
#matplotlibで描画
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.imshow(sample)
plt.show()
```

```python
sample.resize(size=(200, 300))
```

```python
#Pillowの中で用意されているフィルターを使用
sample.resize(size=(200, 300),resample=Image.Resampling.NEAREST)
```

```python
sample.rotate(45)
```

```python
sample.rotate(15, expand=True)
```

```python
#画像の切り出し
sample.crop((0, 0, 540, 400)) #左上と右下の座標をタプルで渡す。左上の座標が(0, 0)
```

```python
crop_img = sample.crop((0, 0, 540, 400))
crop_img.save('crop_img.png')
```

```python
#画像はピクセルごとの色の数値データの集まり
#そのためnumpyに変換も可能
import numpy as np
num_img = np.array(sample)
num_img.shape
```

```python
num_img[0, 0] #ピクセルごとにデータを取得
#データは RGBA を表し、Aは透明度。0~255の値をとり、すべて最大であるので白色であるとわかる
```

```python
num_img[100, 100]
```
