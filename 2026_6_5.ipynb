scikit-learnの復習

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'A': [1, np.nan, 3, 4, 5],
        'B': [6, 7, 8, np.nan, 10],
        'C': [11, 12, 13, 14, 15]
    }
)
df
```

```python
df.isnull()  #欠損値かどうかを確かめる
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')  #平均値で欠損値を埋めるインスタンス
imp.fit_transform(df)
```

```python
df = pd.DataFrame(
    {
        'A': [1, 2, 3, 4, 5],
        'B': ['a', 'b', 'a', 'b', 'c']  #カテゴリ変数
    }
)
df
```

```python
#カテゴリ変数のエンコーディング
#a=0, b=1, c=2　のようにカテゴリを数字に変換する
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df.loc[:, 'B'])
```

```python
le.classes_
```

```python
df = pd.DataFrame(
    {
        'A': [1, 2, 3, 4,5],
        'B': [100, 200, 400, 500, 800]
    }
)
df
```

```python
#データの標準化
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
stdsc.fit_transform(df)
```
